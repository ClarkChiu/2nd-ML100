```python
import pandas as pd
train_data = pd.read_csv('data/data-science-london-scikit-learn/train.csv')
train_target = pd.read_csv('data/data-science-london-scikit-learn/trainLabels.csv')
# train_data = pd.merge(train, train_target, left_index=True, right_index=True)
train_data.head(10)
```

```python
train_data.describe()
```

```python
print(train_data.shape)
```

```python
print(train_target.shape)
```

```python
from sklearn import metrics
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(train_data, train_target.values.ravel())
rfc = RandomForestClassifier(n_estimators=100)
rfc.fit(x_train, y_train)
y_pred = rfc.predict(x_test)
print(f'Accuracy: {metrics.accuracy_score(y_test, y_test)}')
```

```python
train_test = pd.read_csv('data/data-science-london-scikit-learn/test.csv', header=None)
train_test.head()
```

```python
train_test.shape
```

```python
test_pred = rfc.predict(train_test)
test_pred_df = pd.DataFrame(test_pred, columns=['Solution'])
test_pred_df.insert(0, 'Id', range(1, 9001))
test_pred_df.tail()
```

```python
test_pred_df.to_csv('data/data-science-london-scikit-learn/pred.csv', index=False)
```
